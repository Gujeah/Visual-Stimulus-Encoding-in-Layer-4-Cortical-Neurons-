```python

frame_unique = sorted(list(stim_table_composed.frame.unique()))


num_stim_conditions = len(frame_unique)
print(f"\n unique stimulus conditions: {num_stim_conditions}")

# indices for pivot table reindexing to ensure consistent shape
full_columns = pd.Index(frame_unique, name='Frames')

# storing one dimention flattened tuning curves
all_neurons_flat_tuning_curves = []

#
num_neurons = dff.shape[0]
print(f"{num_neurons} neurons to extract their tuning profiles")

for neuron_idx in range(num_neurons):
    neuron_dff = dff[neuron_idx, :]

    # mean dF/F for each stimulus presentation for each current neuron
    current_neuron_stim_data = stim_table_composed.copy()
    current_neuron_stim_data['mean_dff'] = [
        np.mean(neuron_dff[int(start):int(end)])
        for start, end in zip(current_neuron_stim_data['start'], current_neuron_stim_data['end'])
    ]

    # tuning curve pivot table for each current neuron
    tuning_curve_pivot = current_neuron_stim_data.pivot_table(
        values='mean_dff',
        columns='frame',
        aggfunc='mean'
    )

    # reindexing to look cool you get what i mean
    tuning_curve_pivot = tuning_curve_pivot.reindex(columns=full_columns)

    # Flattening the pivot table data for this neuron into a 1D array
    pivot_flat = np.array(tuning_curve_pivot.values).flatten()

    # Appending the flattened neuron tuning curve to the list that we created  above
    all_neurons_flat_tuning_curves.append(pivot_flat)


# creating a Matrix of  (Neurons x Stimulus Conditions) 
response_matrix = np.array(all_neurons_flat_tuning_curves)
response_matrix.shape

```

```python
response_matrix



mean = np.mean(response_matrix, axis=1)
mean = mean[:,np.newaxis]

std = np.std(response_matrix, axis=1)
std = std[:,np.newaxis]


m = (response_matrix - mean)/std
print(m.shape)

```

## Running PCA for natural scenes 

```python
response_matrix_natural_scenes
```

```python
Xplan1=response_matrix_natural_scenes.T
Xplan1.shape
mu = np.mean(Xplan1,axis=0)
n  = Xplan1.shape[0]
p  = Xplan1.shape[1]

# now compute the data covariance
Xplan_centered = Xplan1-mu                         # don't forget to center your data first!
cov = (1/n) * Xplan_centered.T @ Xplan_centered   # now you can compute the covariance

# perform eigendecomposition of the data covariance
D, U = np.linalg.eig(cov) # D = vector of eigenvalues, U = eigenvectors matrix

# make sure the eigenvalues are sorted (in descending order)
indx = np.argsort(D)[::-1]
D    = D[indx]

# arrange the eigenvectors according to the magnitude of the eigenvalues
U = U[:,indx]
#plot the square-root eigenvalue spectrum
fig = plt.figure()
plt.plot(np.sqrt(D),'-o',color='black')
plt.xlabel('dimensions')
plt.ylabel('square-rooted eigenvalues')
plt.title('Square-root eigenvalue spectrum for natural scenes')
plt.show()




```

```python

U = np.real(U)
fig = plt.figure()
ax  = fig.add_subplot(111, projection='3d')

X = Xplan_centered
ax.scatter(X@U[:,0],X@U[:,1],X@U[:,2],c='blue', marker='o')

```

```python
fig = plt.figure(figsize=(15,5))
plt.subplot(1,3,1)

U3  = U[:,0]
b   = plt.imshow(U3[np.newaxis,:].T, aspect='auto')
fig.colorbar(b)
plt.xticks([0],[1])
plt.xlabel('Principal component #')
plt.ylabel('Neuron #')
plt.title('Heat map of 1st principal component directions')
plt.subplot(1,3,2)
U3  = U[:,1]
b   = plt.imshow(U3[np.newaxis,:].T, aspect='auto')
fig.colorbar(b)
plt.xticks([0],[1])
plt.xlabel('Principal component #')
plt.ylabel('Neuron #')
plt.title('Heat map of 2nd principal component directions')
plt.subplot(1,3,3)
U3  = U[:,2]
b   = plt.imshow(U3[np.newaxis,:].T, aspect='auto')
fig.colorbar(b)
plt.xticks([0],[1])
plt.xlabel('Principal component #')
plt.ylabel('Neuron #')
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')


plt.title('Heat map of 3rd principal component directions')

plt.tight_layout()
plt.savefig(r'C:\Users\Ethel\Desktop\project_pictures\natural stimulus\ Heat map of 3rd principal component directionsfor natural.png', format='png')
plt.show()
plt.show()
```
