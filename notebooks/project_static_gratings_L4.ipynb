## Multiplying the neuron by its transpose  X.X.T 


```python

orientation_unique = sorted(list(stim_table_composed.orientation.unique()))
spatial_frequency_unique = sorted(list(stim_table_composed.spatial_frequency.unique()))

num_stim_conditions = len(orientation_unique) * len(spatial_frequency_unique)
print(f"\n unique stimulus conditions: {num_stim_conditions}")

# indices for pivot table reindexing to ensure consistent shape
full_index = pd.Index(orientation_unique, name='orientation')
full_columns = pd.Index(spatial_frequency_unique, name='spatial_frequency')

# storing one dimention flattened tuning curves
all_neurons_flat_tuning_curves = []

#
num_neurons = dff.shape[0]
print(f"\nProcessing {num_neurons} neurons to extract their tuning profiles")

for neuron_idx in range(num_neurons):
    neuron_dff = dff[neuron_idx, :]

    # mean dF/F for each stimulus presentation for each current neuron
    current_neuron_stim_data = stim_table_composed.copy()
    current_neuron_stim_data['mean_dff'] = [
        np.mean(neuron_dff[int(start):int(end)])
        for start, end in zip(current_neuron_stim_data['start'], current_neuron_stim_data['end'])
    ]

    # tuning curve pivot table for each current neuron
    tuning_curve_pivot = current_neuron_stim_data.pivot_table(
        values='mean_dff',
        index='orientation',
        columns='spatial_frequency',
        aggfunc='mean'
    )

    # reindexing to look cool you get what i mean
    tuning_curve_pivot = tuning_curve_pivot.reindex(index=full_index, columns=full_columns).fillna(0)

    # Flattening the pivot table data for this neuron into a 1D array
    pivot_flat = np.array(tuning_curve_pivot.values).flatten()
    
    # Appending the flattened neuron tuning curve to the list that we created  above
    all_neurons_flat_tuning_curves.append(pivot_flat)


# creating a Matrix of  (Neurons x Stimulus Conditions) 
response_matrix = np.array(all_neurons_flat_tuning_curves)
response_matrix.shape

```

## Dimension Reduction using PCA

```python
Xplan=response_matrix.T
Xplan.shape
```

```python
mu = np.mean(Xplan,axis=0)
n  = Xplan.shape[0]
p  = Xplan.shape[1]

# now compute the data covariance
Xplan_centered = Xplan-mu                         # don't forget to center your data first!
cov = (1/n) * Xplan_centered.T @ Xplan_centered   # now you can compute the covariance

# perform eigendecomposition of the data covariance
D, U = np.linalg.eig(cov) # D = vector of eigenvalues, U = eigenvectors matrix

# make sure the eigenvalues are sorted (in descending order)
indx = np.argsort(D)[::-1]
D    = D[indx]

# arrange the eigenvectors according to the magnitude of the eigenvalues
U = U[:,indx]
```

```python
np.linalg.matrix_rank(Xplan)
```

```python
#plot the square-root eigenvalue spectrum
fig = plt.figure()
plt.plot(np.sqrt(D),'-o',color='black')
plt.xlabel('dimensions')
plt.ylabel('square-rooted eigenvalues')
plt.title('Square-root eigenvalue spectrum')
plt.show()
```

```python
percent_var = 100 * np.sum(D[0:3])/np.sum(D)
print('percent variance explained = ', percent_var)
```

## trying to understand the concept of neural population's homogeneity, dont know what it is but i found it on internet😎😎😎 <br>
Steps to do this <br>
1. Extracting all unique pairwise correlations from your correlation_df.<br>

2. Excluding the self-correlations (where a neuron correlates with itself, which is always 1).<br>

3. Excluding duplicate pairs (e.g., the correlation between Neuron A and Neuron B is the same as Neuron B and Neuron A, so we only need to list it once).<br>

4. Sorting them in descending order and taking the top N.<br>

```python
stacked_corr=correlation_df.stack()
stacked_corr
```

```python
flat_correlations = stacked_corr.reset_index()
flat_correlations.columns = ['neuron1', 'neuron2', 'correlation']
```

```python
#removing the correlation of a neuron by itself 
non_self_correlations =flat_correlations[flat_correlations['neuron1']!=flat_correlations['neuron2']]
# our matrixsymmetric (corr(A,B) == corr(B,A)) so the opposite is true
unique_pairs_correlations = non_self_correlations[non_self_correlations['neuron1'] < non_self_correlations['neuron2']]
top_correlated_pairs = unique_pairs_correlations.sort_values(by='correlation', ascending=False)
```

```python
matrix1=np.triu(loaded_matrix_natural_scenes, k=1)
matrix2=np.triu(loaded_matrix_static_gratings, k=1)
```

```python
from scipy.stats import spearmanr
from scipy.stats import pearsonr
```

```python
def spearman_correlation_matrices(matrix1, matrix2):
   
    return  pearsonr(matrix1.flatten(), matrix2.flatten())
```

```python
correlation_of_correlations=spearman_correlation_matrices(matrix1, matrix2)
correlation_of_correlations
```

## shuffling control

```python
def spearman_correlation_matrices(matrix1, matrix2):
    matrix1_flattened = matrix1.flatten()
    np.random.shuffle(matrix1_flattened)
    print(type(matrix1_flattened))
    return  pearsonr(matrix1_flattened, matrix2.flatten())
```

```python
correlation_of_correlations=spearman_correlation_matrices(matrix1, matrix2)
```
